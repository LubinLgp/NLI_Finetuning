```python
import pandas as pd
from sklearn.metrics import f1_score

# Utilitaires communs
BASE_DIR = "../results"  # ce notebook est déjà dans results/

LABEL_ENTAIL = "Entailment"


def macro_f1(df):
    """Macro F1 binaire/3-classes (labels tels que dans les CSV)."""
    y_true = df["gold"]
    y_pred = df["prediction"]
    return f1_score(y_true, y_pred, average="macro")


def load_split(path):
    """Charge un CSV et renvoie (global, pol, medical)."""
    df = pd.read_csv(path)
    df_pol = df[df["statement_type"] == "pol"]
    df_med = df[df["statement_type"] == "medical"]
    return df, df_pol, df_med


def format_percent(x: float) -> str:
    return f"{100 * x:.1f}"
```

```python
# 1) NLI Baseline

path = "Prompt_NLI/eval_nli_combined_baseline.csv"
df, df_pol, df_med = load_split(path)

f1_global = macro_f1(df)
f1_pol = macro_f1(df_pol)
f1_med = macro_f1(df_med)

print("NLI Baseline — macro F1 (%)")
print("  Global :", format_percent(f1_global))
print("  POL    :", format_percent(f1_pol))
print("  MEDICAL:", format_percent(f1_med))
```
